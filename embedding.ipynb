### Word2Vec Embedding

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the Vocabulary size
vocab_size=10000
```

```python
### One Hot Representation
one_hot_representations = [one_hot(words, vocab_size) for words in sent]
one_hot_representations
```

```python
## Word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_len=8
embedded_docs = pad_sequences(one_hot_representations, padding='pre', maxlen=sent_len)
embedded_docs
```

```python
## Feature Representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_len))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
